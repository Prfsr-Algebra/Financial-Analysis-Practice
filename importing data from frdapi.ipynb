#### Putting the series IDs you want to use in alist

```python
series_ids = ['DGS1', 'DGS2', 'DGS3']
```

#### Creating Function for getting data for a single series:

```python
def get_series_data(series_id, start, end):
    data = fred.get_series(series_id, observation_start = start, observation_end = end)
    return data
```

#### Getting data for all series in the series_ids list

```python
start = "2001-01-01"
end = "2024-12-31"
yields_dict = {series_id: get_series_data(series_id, start, end) for series_id in series_ids}
```

#### Converting to a dataframe

```python
yields_df = pd.DataFrame(yields_dict)
yields_df.head()
```
